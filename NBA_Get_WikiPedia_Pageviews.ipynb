```
from mwviews.api import PageviewsClient

agent = 'Mozilla/5.0 (X11; CrOS x86_64 10575.58.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36'
p = PageviewsClient(agent)
x = p.article_views('en.wikipedia', ['Selfie', 'Cat'], granularity='daily', start='20160201', end='20160201')
```

```
df1 = pd.read_csv("drive/NBA_Data_Hackathon/nba1617_gt.csv")
```

```
teams = list(df1.Home.unique())
```

```
df1.head()
```

```
seasons = pd.date_range("2016-01-01", "2018-09-09").to_series()

for day in seasons[:10]:
  start = day.strftime('%Y%m%d')
  print(start)
```

```
columns = ['date', 'team', 'wp_pageviews']
dft = pd.DataFrame(columns=columns)

for day in seasons:
  
    game_date = day.strftime('%Y%m%d')    
    print("Processing date: {0}".format(game_date))
    x = p.article_views('en.wikipedia', teams, granularity='daily', start=game_date, end=game_date)
    dfx = pd.DataFrame.from_dict(x).T.reset_index()
    
    for team in teams:
      team_key = team.replace(' ','_')
      team_views = dfx[team_key].iloc[0]
      dft = dft.append({'date': day,
                        'team': team,
                        'wp_pageviews': team_views}, ignore_index=True)
```
